```python
import struct 
 
def read_bin(file_name):
    """
    function: read a bin file, return the tuple of the content in file
    """
    with open(file_name, "rb") as f:
        f_content = f.read()
        content = struct.unpack("B" * len(f_content), f_content)
        f.close()
    return content
```
